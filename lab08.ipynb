```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

# Part 1

```python
'''
Loading data from file "droptower_vdata.txt"
t = time (0-10 s)
h = position (height)
v = velocity
a = acceleration

'''
t, v = np.loadtxt("droptower_vdata.txt", unpack=True)
h = integrate.cumtrapz(v, t, initial=0)
a = np.diff(v)
a = np.append(a, a[len(a)-1]) #repeat the last element to match dimensions
```

```python
'''
Graphing data

'''

f, (ht, vt, at) = plt.subplots(3, 1)
ht.set_title('Height', fontsize=20)
ht.plot(t, h)
vt.set_title('Velocity', fontsize=20)
vt.plot(t, v)
at.set_title('Acceleration', fontsize=20)
at.plot(t, a)
plt.tight_layout()
plt.subplots_adjust(top=1.5, bottom=0.1, hspace=0.5)
plt.suptitle('Drop Tower Data', fontsize=25, y=1.7)
```

# Part 2

```python
a_positive = a[a >= 0]
print(a_positive)
avg_up_accel = np.average(a_positive)
print("Average upwards acceleration: " + str(avg_up_accel))
```
